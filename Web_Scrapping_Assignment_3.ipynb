```python
from bs4 import BeautifulSoup
import requests
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.amazon.in/")
time.sleep(5)
```

```python
search_input = driver.find_element(By.ID,"twotabsearchtextbox")
search_product = input("Enter the product to search: ")
search_input.send_keys(search_product,'s')
```

```python
Search=driver.find_element(By.ID,"nav-search-submit-button")
Search.click()
time.sleep(3)
```

```python
Product_url=[]
```

```python
start=0
end=3 
try:
    for page in range(start,end):  
        url = driver.find_elements(By.XPATH, '//a[contains(@class, "a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal")]')
        for i in url:
            Product_url.append(i.get_attribute('href'))
        Next = driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
        Next.click()
except NoSuchElementException :
        Product_url.append('No url')
```

```python
Brand=[] 
Name_of_product=[] 
Price=[] 
Return_Exchange=[] 
Expected_Delivery=[] 
Availability=[]

for url in Product_url[0:50]: 
    driver.get(url) 
    time.sleep(1)
    try:
        brands = driver.find_element(By.XPATH,'//div[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span') 
        Brand.append(brands.text) 
    except NoSuchElementException : 
        Brand.append('-')

    try:
        Product = driver.find_element(By.XPATH,'//span[@id="productTitle"]')
        Name_of_product.append(Product.text)
    except NoSuchElementException :
        Name_of_product.append('-')

    try:
        Prices = driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        Price.append(Prices.text)
    except NoSuchElementException :
        Price.append('-')

    try:
        delivery = driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[7]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[3]/div/div[2]/div[9]/div[1]/div/div/div/span/span')
        Expected_Delivery.append(delivery.text)
    except NoSuchElementException :
        Expected_Delivery.append('-')

    try:
        re = driver.find_element(By.XPATH,'//div[@id="RETURNS_POLICY"]')
        Return_Exchange.append(re.text)
    except NoSuchElementException :
        Return_Exchange.append('-')

    try:
        Avl = driver.find_element(By.XPATH,'//div[@id="availability"]')
        Availability.append(Avl.text)
    except NoSuchElementException :
        Availability.append('-')
    
    except InvalidArgumentException as e:
        print("Invalid argument error:", e)
```

Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
hostel_name=[] 
Rating=[] 
total_review=[] 
overall_review=[] 
facilities=[]
prop_des=[]

for url in hostel_url: 
    driver.get(url) 
    time.sleep(1)
    try:
        hosname = driver.find_element(By.XPATH,'//div[@class="title-2"]') 
        hostel_name.append(hosname.text) 
    except NoSuchElementException : 
        hostel_name.append('-')

    try:
        Rat = driver.find_element(By.XPATH,'//div[@class="score orange big"]')
        Rating.append(Rat.text)
    except NoSuchElementException :
        Rating.append('-')

    try:
        tr = driver.find_element(By.XPATH,'//div[@class="reviews"]')
        total_review.append(tr.text)
    except NoSuchElementException :
        total_review.append('-')

    try:
        ore = driver.find_element(By.XPATH,'//div[@class="keyword"]')
        overall_review.append(ore.text)
    except NoSuchElementException :
        overall_review.append('-')
        
    try:
        fac = driver.find_element(By.XPATH,'//ul[@class="facility-sections"]')
        facilities.append(fac.text)
    except NoSuchElementException :
        facilities.append('-')
        
    try:
        prde = driver.find_element(By.XPATH,'//div[@class="content collapse-content"]')
        prop_des.append(prde.text)
    except NoSuchElementException :
        prop_des.append('-')
```

```python
Price=[]
location=[]
pp_tags = driver.find_elements(By.CLASS_NAME,'prices')
for i in pp_tags[0:50]:
    price = i.text
    Price.append(price)
loc_tags = driver.find_elements(By.XPATH,'//span[@class="description"]/span[1]')
for i in loc_tags[0:50]:
    loca = i.text
    location.append(loca)
```

```python
Price
```

```python
Private_Price = [item.split('\n')[1].strip() for item in Price[1:50]]
print(Private_Price)
```

```python
Dorms_Price = [item.split('\n')[2].strip() for item in Price[1:50]]
print(Dorms_Price)
```

```python
pp = [s.replace('Dorms From', '-') for s in Private_Price]
print(pp)
dp = [s.replace('Dorms From', '-') for s in Dorms_Price]
print(dp)
```

```python
import pandas as pd
df9=pd.DataFrame({'Hostel name': hostel_name,'Location':location,'Rating': Rating,'Total review': total_review,'Overall review': overall_review,'Private room Price': pp[0:30],'Dorms price': dp[0:30],'Facilities':facilities,'Property Description':prop_des})
df9
```

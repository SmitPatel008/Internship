# Prediction of earning less than 50K or more than 50K

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
x = df1.iloc[:,:4].join(df1.iloc[:, 5:6]).join(df1.iloc[:, 7:11])
```

```python
y=df1.iloc[:,-1]
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st=StandardScaler()
x=st.fit_transform(x)
x
```

we applying scaller to our data for better perfomance of our model.

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=40)
```

```python
lg = LogisticRegression()
lg.fit(x_train, y_train)
pred=lg.predict(x_test)
accuracy_score(y_test,pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtc.score(x_train,y_train)
preddtc=dtc.predict(x_test)
print(accuracy_score(y_test,preddtc))
print(confusion_matrix(y_test,preddtc))
print(classification_report(y_test,preddtc))
```

DecisionTreeClassifier model has accuracy score is 80%.

```python
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
knn.score(x_train,y_train)
predknn=knn.predict(x_test)
print(accuracy_score(y_test,predknn))
print(confusion_matrix(y_test,preddtc))
print(classification_report(y_test,preddtc))
```

KNeighborsClassifier model accuracy score is 81%.

```python
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier,AdaBoostClassifier, GradientBoostingClassifier, BaggingClassifier
```

```python
RFC = RandomForestClassifier()
RFC.fit(x_train, y_train)
predRFC = RFC.predict(x_test)
print(accuracy_score(y_test, predRFC))
print(confusion_matrix(y_test,predRFC))
print(classification_report(y_test,predRFC))
```

RandomForestClassifier model accuracy score is 83%.

```python
ad=AdaBoostClassifier()
ad.fit(x_train,y_train)
ad_pred=ad.predict(x_test)
print(accuracy_score(y_test,ad_pred))
print(confusion_matrix(y_test,ad_pred))
print(classification_report(y_test,ad_pred))
```

AdaBoostClassifier model accuracy score is 85%.

```python
GB = GradientBoostingClassifier()
GB.fit(x_train, y_train)
predGB = GB.predict(x_test)
print(accuracy_score(y_test, predGB))
print(confusion_matrix(y_test,predGB))
print(classification_report(y_test,predGB))
```

GradientBoostingClassifier model accuracy score is 85%.

```python
print(accuracy_score(y_test, predGB))
```

```python
BC = BaggingClassifier()
BC.fit(x_train, y_train)
predBC = BC.predict(x_test)
print(accuracy_score(y_test, predBC))
print(confusion_matrix(y_test,predBC))
print(classification_report(y_test,predBC))
```

BaggingClassifier model accuracy score is 81%.

```python
ETC = ExtraTreesClassifier()
ETC.fit(x_train, y_train)
predETC = ETC.predict(x_test)
print(accuracy_score(y_test, predETC))
print(confusion_matrix(y_test,predETC))
print(classification_report(y_test,predETC))
```

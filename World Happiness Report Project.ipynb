```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df1=df.iloc[:,4:]
df2=df.iloc[:,:3]
```

```python
df5
```

```python
x=df5.iloc[:,0:-1]
y=df5.iloc[:,-1]
```

```python
y
```

```python
lm=LinearRegression()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df5['Country'] = le.fit_transform(df5['Country'])
df5['Region'] = le.fit_transform(df5['Region'])
```

```python
df5
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=42)
```

```python
lm.fit(x_train,y_train)
```

```python
lm.score(x_train,y_train)
```

Our Model Accuracy is very high which is 99.99%

```python
pred=lm.predict(x_test)
```

```python
print('Predicted score:',pred)
print('actual score',y_test)
```

As here we can see our predicted score is highy match with actual score as our model accuracy is 99.99%.
